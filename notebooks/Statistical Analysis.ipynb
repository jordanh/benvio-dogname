```python
import pandas
import numpy as np
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels

%matplotlib inline
%config InlineBackend.figure_format = 'png'
```

Experiment 1 (Objective Humor)
==============================

Data Import from CSV
--------------------

```python
exp1DF = pandas.read_csv("../data/results-exp1.csv")
```

Exploratory Scatter Plot
------------------------

We color dog/name combinations that were reported as "funny" yellow. At first blush, it appears that funny dog/name combinations are randomly assorted through the data.

```python
g = sns.lmplot("Violation Sample Value", "Benign Sample Value", exp1DF, hue="Funny Sample Actual",
            palette=np.array(sns.color_palette())[[4,0]], markers=['o','.'],
            fit_reg=False, x_jitter=0.25, y_jitter=0.25)
g.set(xlim=(-2.5, 2.5), ylim=(-2.5, 2.5))
```

Linear Regression Plot
----------------------

N.B. that this is a simple linear regression. We used JMP for our analysis which performed a two-way regression and calculated much tighter confidence intervals around the fit lines as more of the data was used.

```python
g = sns.lmplot("Benign Sample Value", "Funny Sample Value", exp1DF,
           hue="Violation Sample Value",  col="Violation Sample Value",
           x_jitter=.25, y_jitter=.25)
g.set(xlim=(-2.5, 2.5), ylim=(-2.5, 2.5))
```

```python
g = sns.lmplot("Benign Sample Value", "Funny Sample Value",
           exp1DF[(exp1DF["Violation Sample Value"] == -2) | 
                  (exp1DF["Violation Sample Value"] == 0)  |
                  (exp1DF["Violation Sample Value"] == 2)],
           hue="Violation Sample Value",  col="Violation Sample Value",
           palette=np.array(sns.color_palette())[[0,2,4]],
           x_jitter=.25, y_jitter=.25)
g.set(xlim=(-2.5, 2.5), ylim=(-2.5, 2.5))
```

Interact Plot
-------------

The below interact plot more clearly shows that as a dog/name combination becomes more and more violating it must also be increasingly benign in order to be considered funny.

```python
g = sns.interactplot("Violation Sample Value", "Benign Sample Value", "Funny Sample Value", exp1DF,
                 cmap="coolwarm", filled=True, levels=5);
```

Experiment 2 (Subjecive Humor)
==============================

Data Import from CSV
--------------------

```python
exp2DF = pandas.read_csv("../data/results-exp2.csv")
```

Linear Regression Plot
----------------------

The subjective experiment more clearly shows the correlations for variously dog/name levels of violation. Humor is, and always will be, a highly subjective experience.

Especially refer to **Violation Sample Value = 2** where funniness is clearly and directly correlated to how benign a dog/name combination is to the participant.

```python
g = sns.lmplot("Benign Sample Value", "Funny Sample Value", exp2DF,
           hue="Violation Sample Value",  col="Violation Sample Value",
           x_jitter=.25, y_jitter=.25)
g.set(xlim=(-2.5, 2.5), ylim=(-2.5, 2.5))
```

Interact Plot
